```python
import pandas as pd
import glob
from sqlalchemy import create_engine

# Path to folder containing files
path = r"C:\_SDATA\__Pers\Courses\20250521_NTUC_Associate_Data_Analyst\Capstone Project -Retail Business analysis\Dataset\Files\*.csv"

# Get all file names
files = glob.glob(path)

# Read and append
df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

```

```python
df.head()
```

```python
df.info()
```

```python
df['discount_price'] = pd.to_numeric(
    df['discount_price'].astype(str).str.replace('[^0-9.]', '', regex=True),
    errors='coerce'
)
```

```python
df['actual_price'] = pd.to_numeric(
    df['actual_price'].astype(str).str.replace('[^0-9.]', '', regex=True),
    errors='coerce'
)
```

```python
df['ratings'] = pd.to_numeric(
    df['ratings'],
    errors='coerce'
)

df = df[df['ratings'].between(0, 5)]
```

```python
df['no_of_ratings'] = pd.to_numeric(df['no_of_ratings'], errors='coerce')
```

```python
threshold = df['actual_price'].quantile(0.95)

df['actual_price'] = df['actual_price'].apply(
    lambda x: x / 100 if x > threshold else x
)

df['discount_price'] = df['discount_price'].apply(
    lambda x: x / 100 if x > threshold else x
)
```

```python
df.describe()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.describe(include = 'all')
```

```python
df.isnull().sum()
```

```python
df['no_of_ratings'] = df['no_of_ratings'].fillna(
    df.groupby(['main_category', 'sub_category'])
        ['no_of_ratings'].transform('median').round()).astype('Int64')
```
